```python
#Бинарный поиск
def binary_search(data,val): #итеративная реализация бинарного поиска
    data.sort()
    first=0
    last=len(data)-1
    index=-1
    while (first<=last) and (index==-1):
        mid = (first+last)//2
        if data[mid]==val:
            index=mid
        else:
            if val<data[mid]:
                last=mid-1
            else:
                first=mid+1
    if (index>=0):
        return index+1
    else:
        return "Value not found"
    
def binary_search_delete(data,val): #находим нужный элемент с помощью бинарного поиска и удаляем его по индексу
    index=binary_search(data,val)
    if index=="Value not found":
        return index
    else:
        data.pop(index-1)
        return data

def binary_search_append(data,val): #добавление элемента
    data.sort()
    if data[-1]<val:     #если добавляемый элемент больше максимального в исходном списке, добавляем его в конец
        data.append(val)
        return data
    if data[0]>val:      #если добавляемый элемент меньше минимального в исходном списке, добавляем его в начало
        data.insert(0,val)
        return data
    #в противном случае при помощи бинарного поиска определяем место вставки нашего элемента в отсортированном списке
    first=0
    last=len(data)-1
    while (first<=last): 
        mid = (first+last)//2
        if data[mid]<=val and data[mid+1]>=val:
            data.insert(mid+1,val)
            return data
        else:
            if val<data[mid]:
                last=mid-1
            else:
                first=mid+1
data1=copy.deepcopy(data)
start_time = time.time()
print(binary_search(data1,20))
binary_search_time=round((time.time() - start_time)*1000)
data1.sort()
print(data1)
print(binary_search_delete(data1,36))
print(binary_search_append(data1,85))
print("--- {0} ms ---".format(binary_search_time))
```

```python
#Бинарное дерево
class Node:
    def __init__(self,x):
        self.left = None
        self.right = None
        self.data = x
        
class Tree:
    root = None
    
    def __init__(self):
        root = None
        
    def insert(self,x):
        if self.root is None:
            self.root = Node(x)
            return
        if x<self.root.data:
            if self.root.left is None:
                self.root.left = Tree()
            self.root.left.insert(x)
        else:
            if self.root.right is None:
                self.root.right = Tree()
            self.root.right.insert(x)
            
    def find(self, x):
        if self.root is None:
            return None
        if x == self.root.data:
            return self.root
        if x<self.root.data:
            if self.root.left is None:
                return None
            return self.root.left.find(x)
        else:
            if self.root.right is None:
                return None
            return self.root.right.find(x)
        
    def print(self):
        if self.root is None:
            return 
        if self.root.left:
            self.root.left.print()
        print (self.root.data, end = ' ')
        if self.root.right:
            self.root.right.print()
            
    def minValueNode(self):
        current=self.root.data
        while(self.root.left is not None):
            current = self.root.left
        return current
            
    def deleteNode(self, x):
        if self.root is None:
            return None
        
        if x < self.root.data:
            self.root.left = self.root.left.deleteNode(x)
        elif(x > self.root.data):
            self.root.right = self.root.right.deleteNode(x)
        else:
            if self.root.left is None:
                temp = self.root.right
                self.root = None
                return temp
            elif self.root.right is None:
                temp = self.root.left
                self.root = None
                return temp
            temp = self.root.right.minValueNode()
            self.root.data = temp
            self.root.right = self.root.right.deleteNode(temp)
        return self.root
```

```python
#Интерполяционный
def interpolation_search(data, val):
    data.sort()
    low=0
    high=len(data)-1
    while low<=high and val>=data[low] and val<=data[high]:
        index = low + int(((float(high-low)/(data[high]-data[low])) * (val-data[low])))
        if data[index]==val:
            return index+1
        if data[index]<val:
            low=index+1
        else:
            high=index-1
    return None

def interpolation_search_delete(data,val):
    i = interpolation_search(data, val) 
    if i is not None:
        data.pop(i-1)
    return data

def interpolation_search_append(data, val):
    data.sort()
    if data[-1]<val:     #если добавляемый элемент больше максимального в исходном списке, добавляем его в конец
        data.append(val)
        return data
    if data[0]>val:      #если добавляемый элемент меньше минимального в исходном списке, добавляем его в начало
        data.insert(0,val)
        return data
    #в противном случае при помощи интерполяционного поиска определяем место вставки нашего элемента в отсортированном списке
    low=0
    high=len(data)-1
    while low<=high and val>=data[low] and val<=data[high]:
        index = low + int(((float(high-low)/(data[high]-data[low])) * (val-data[low])))
        if data[index]<=val and data[index+1]>=val:
            data.insert(index+1,val)
            return data
        elif data[index]<val:
            low=index+1
        else:
            high=index-1
data3=copy.deepcopy(data)
start_time=time.time()
print(interpolation_search(data3,65))
interpolation_search_time=round((time.time() - start_time)*1000)
data3.sort()
print(data3)
print(interpolation_search_delete(data3,1))
print(interpolation_search_append(data3,65))
print("--- {0} ms ---".format(interpolation_search_time))
```

```python
name_list = ["Terry", "Jose", "Betty", "Debra", "Larry", "Robert", "Sherri", "Gary", "Donna", "Carol",
            "Leroy", "Norma", "Robin", "David", "Tammy", "Julia", "Floyd", "Velma", "Sarah", "Marcia",]

def return_sum(name):
    sum = 0
    for el in name: sum += ord(el)
    print(sum)

for name in name_list:
    return_sum(name)
```

```python
#Простое рехэширование

name_hash_list = [0 for i in range(len(name_list))]
def simple_hash(name, size, p = 0):
    result = 0
    for el in name:
        result += ord(el)
    return (result + p) % size

for name in name_list:
    i = simple_hash(name, len(name_list))
    p = 1
    while name_hash_list[i] != 0:
        i = simple_hash(name, len(name_list), p)
        p += 1
    name_hash_list[i] = name

print(name_hash_list)
for name in name_list:
    i = simple_hash(name, len(name_list))
    p = 1
    while name_hash_list[i] != name:
        i = simple_hash(name, len(name_list), p)
        p += 1
    print(name_hash_list[i], i)

for name in name_list: #вывод для демонстрации кол-ва коллизий
    print(name, simple_hash(name, len(name_list)), name == name_hash_list[simple_hash(name, len(name_list))])
```

```python
#Рехэширование с помощью псевдослучайных чисел
name_hash_list = [0 for i in range(len(name_list))]

def generate_random_num(name, size, p):
    a = 3
    b = 7
    return ((p ** b) // (size ** a)) + len(name) ** (b - a)

def random_hash(name, size, p = 0):
    result = 0
    for el in name:
        result += ord(el)
    return (result + generate_random_num(name, size, p)) % size

for name in name_list:
    i = random_hash(name, len(name_list))
    p = 1
    while name_hash_list[i] != 0:
        i = random_hash(name, len(name_list), p)
        p += 1
    name_hash_list[i] = name

print(name_hash_list)
#вывод имени и его позиции в списке
for name in name_list:
    i = random_hash(name, len(name_list))
    p = 1
    while name_hash_list[i] != name:
        i = random_hash(name, len(name_list), p)
        p += 1
    print(name_hash_list[i], i)

for name in name_list: #вывод для демонстрации кол-ва коллизий
    print(name, random_hash(name, len(name_list)), name == name_hash_list[random_hash(name, len(name_list))])
```

```python
#Метод цепочек
class Node():
    
    def __init__(self, data = None):
        self.data = data
        self.next = None
    
    def add(self, data):
        if self.data is None:
            self.data = data
        elif self.next is None:
            self.next = Node(data)
        else:
            self.next.add(data)
    def print_Node(self):
        if self.data:
            print(self.data, end = ' ')
            if self.next:
                self.next.print_Node()
            else:
                return
name_hash_list = [Node() for i in range(len(name_list))]

def node_hash(name, size):
    result = 0
    for el in name:
        result += ord(el)
    return result % size

for name in name_list:
    i = node_hash(name, len(name_list))
    name_hash_list[i].add(name)

#печать получившегося списка
for i in range(len(name_hash_list)):
    print(i, end = ' ')
    name_hash_list[i].print_Node()
    print()
```

### Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.


```python
def place_queen(board, i, j):
    for k in range(len(board)):
        board[i][k] += 1 #увеличиваем число ферзей, которые бьют эту клетку путем прибавления единицы
        board[k][j] += 1
        if 0 <= i + j - k < len(board):
            board[i + j - k][k] += 1
        if 0 <= i - j + k < len(board):
            board[i - j + k][k] += 1
    board[i][j] = -1 #в клетке с отрицательным значением размещается ферзь

def remove_queen(board, i, j):
    for k in range(len(board)):
        board[i][k] -= 1 #уменьшаем число ферзей, которые бьют эту клетку путем убавления единицы
        board[k][j] -= 1
        if 0 <= i + j - k < len(board):
            board[i + j - k][k] -= 1
        if 0 <= i - j + k < len(board):
            board[i - j + k][k] -= 1
    board[i][j] = 0 #клетка с нулевым значением пригодна для размещения ферзя: там ещё нет ферзя и её никто не бьёт

def print_answer(board):
    res = ""
    for i in range(len(board)):
        for j in range(len(board)):
            if board[i][j] == -1:
                res += chr(97 + i) + str(j + 1) + ' '
    print(res)
                
def __queens_problem(board, j):
    if j == len(board):
        print_answer(board)
        return
    for i in range(len(board)):
        if board[i][j] == 0:
            place_queen(board, i, j)
            __queens_problem(board, j + 1)
            remove_queen(board, i, j)

def queens_problem(n):
    board = [[0 for i in range(n)] for j in range(n)]
    __queens_problem(board, 0)
    
queens_problem(8)
```
